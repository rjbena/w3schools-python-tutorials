```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0,6])
ypoints = np.array([0,250])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
# functions for plot are marker, markersize as ms, markeredgecolor as markeredgecolor, markerfacecolor as mfc, linestyle as ls

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3,8,1,10])

# Can set font with fontdict
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}


plt.xlabel('X axis', fontdict=font2)
plt.ylabel('Y axis',fontdict=font2)
plt.title('Title', fontdict=font1, loc='center')

plt.plot(ypoints,marker='*', markeredgecolor='g', markerfacecolor='r', markersize=10, linestyle='--', color='black',linewidth='12.2')
# can add grid with grid()
plt.grid()
plt.show()
```

```python
# Create a scatter plot.

import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
plt.scatter(x, y, color = "blue",s=sizes,alpha = 0.5) # can color each dot using c = array_of_colors, alpha adjust transparency
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

```python
# Creating bar charts. 

import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y,color='orange',width=.5) #barh() an be used for horizontal bar. height is used for horizontal bars.
```

```python
# Creating a histogram

import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show() 
```

```python
# Creating a pie chart

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, explode = myexplode, shadow=True, colors=mycolors)
plt.legend(title="4 Fruits")
plt.show() 
```
